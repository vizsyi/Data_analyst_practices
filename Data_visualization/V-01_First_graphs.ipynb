### Matplotlib's two interfaces

#### Imperative interface

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(8, 6))
plt.plot(x, x, label = "linear", color = "#38b2a1", linewidth=1.5, linestyle = "-.")
plt.plot(x, np.cos(x), label = "cosine", color = "#aa0074", linewidth=2, linestyle = ":")
plt.plot(x, np.sin(x), label = "sine", color = "#ccbbdd", linewidth=2, linestyle = "--")

plt.xlabel("x label", size = 14)
plt.ylabel("y label", size = 14)
plt.title("Pyplot functions' interface", size = 18)

plt.legend(loc = "upper center")
```

#### Object oriented interface

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, x, label = "linear", color = "#38b2a1", linewidth=1.5, linestyle = "-.")
ax.plot(x, np.cos(x), label = "cosine", color = "#aa0074", linewidth=2, linestyle = ":")
ax.plot(x, np.sin(x), label = "sine", color = "#ccbbdd", linewidth=2, linestyle = "--")

ax.set_xlabel("x label", size = 14)
ax.set_ylabel("y label", size = 14)
ax.set_title("Object oriented interface", size = 18)

ax.legend(["linear2", "cosine2", "sine2"], loc = "lower left")
```

### Marking points in the graph

```python
x = np.linspace(0, 3, 1000)

plt.figure(figsize=(6, 4))
plt.plot(x, x ** 2, x, 2*x)
plt.annotate("x=0", xy=(0, 0), xytext=(0, 2),
             arrowprops=dict(facecolor="black", arrowstyle="->"))
#applying bounding box
plt.annotate("x=2", xy=(2, 4), xytext=(2, 6),
             arrowprops=dict(facecolor="black", arrowstyle="->"),
             bbox=dict(boxstyle="round", alpha=.1))
```

### Horizontal and vertical lines

```python
x = np.linspace(0, 3, 100)
plt.plot(x, x**2, x, 2*x)

plt.axvline(2, color="gray", alpha=.5, linestyle="-.")
plt.axhline(4, color="gray", alpha=.5, linestyle="-.")
```

### Drawings (patches) in the graph

```python
import matplotlib.patches as patches
```

```python
x = np.linspace(0, 3, 100)
plt.plot(x, x**2, x, 2*x)

p1 = patches.Ellipse(xy=(0, 0), width=.15, height=.6, alpha=.3, facecolor="#aaaaff",
                     edgecolor="black", linewidth=1, linestyle="solid")
p2 = patches.Ellipse(xy=(2, 4), width=.15, height=.6, alpha=.3, facecolor="#aaaaff",
                     edgecolor="black", linewidth=1, linestyle="solid")

ax = plt.gca()
ax.add_patch(p1)
ax.add_patch(p2)
```

### Format axes properties

```python
def pi_formatter(x, pos):
    return "${:1.0f}\pi$".format(x / np.pi)
```

```python
from matplotlib.ticker import FuncFormatter

form = FuncFormatter(pi_formatter)
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

plt.plot(x, np.sin(x))
plt.xlim(-7, 7)
plt.ylim(-2, 2)

plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi])
plt.yticks([-1, 0, 1])
plt.grid(True)

plt.xlabel("x", size = 14)
plt.ylabel("sin(x)", size = 14)

plt.gca().xaxis.set_major_formatter(form)
```
